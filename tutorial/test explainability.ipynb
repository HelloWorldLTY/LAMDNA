```python
# model = LightningModel.load_from_checkpoint('./generation_more_test_value_align_qwen2.53b100epoch_3cl_crosscelline/lightning_logs/version_5/checkpoints/epoch=6-step=21056.ckpt', llm_name = "Qwen/Qwen2.5-3B-Instruct",
#         latent_size = 2048, max_token_len=instruct_len)

model = LightningModel.load_from_checkpoint('./generation_more_test_value_align_qwen2.53b100epoch_3cl_crosschr_predexp/lightning_logs/version_2/checkpoints/epoch=18-step=34216-v1.ckpt', llm_name = "Qwen/Qwen2.5-3B-Instruct",
        latent_size = 2048, max_token_len=instruct_len)
```

```python
for param in model.Llama_encoder.parameters():
    param.requires_grad = True
```

```python
token_emb = model.Llama_encoder(item[0].long().view(1,-1).to('cuda') ,return_dict=True, output_hidden_states=True)['hidden_states'][-1]
```

```python
ig = IntegratedGradients(model.forward_explain)
```

```python
tokenizer = AutoTokenizer.from_pretrained("Qwen/Qwen2.5-3B-Instruct")
decode_map = {}
for val in item[0]:
    decode_map[val] = tokenizer.decode(val)
```

```python
list(decode_map.values())
```

```python
from tqdm import tqdm
```

```python
#explain whole sequence:
attr_list = []
for idx_range in tqdm(range(0,200)):
    input1_attr = ig.attribute(token_emb, additional_forward_args=item[1][0:2+idx_range].view(1,-1).to('cuda'), target = item[2][idx_range], n_steps=100, return_convergence_delta=False)
    out_attr = input1_attr.mean(axis=-1).detach().cpu()
    out_attr = out_attr.numpy()
    attr_list.append(out_attr)
```

```python
attr_list = np.array(attr_list)[:,0,:]
```

```python
df = pd.DataFrame(attr_list)
```

```python
df.index = [idx for idx in train_new.values[0][1]]
df.columns = list(decode_map.values())
```
